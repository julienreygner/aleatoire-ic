## Intervalle de confiance dans le modèle gaussien

```python
import numpy as np

# Génération de l'échantillon
n = 15
mu = np.sqrt(2)
sigma = 1

sample = np.random.normal(loc=mu, scale=sigma, size=n)

print("Échantillon :", np.round(sample, 3))

# Calcul de l'intervalle de confiance (niveau 95 %)
phi = 1.96

moyenne = np.mean(sample)
borne_inf = moyenne - phi * (sigma / np.sqrt(n))
borne_sup = moyenne + phi * (sigma / np.sqrt(n))

print(f"\nMoyenne empirique : {moyenne:.3f}")
print(f"Intervalle de confiance : [ {borne_inf:.3f} , {borne_sup:.3f} ]")
```
